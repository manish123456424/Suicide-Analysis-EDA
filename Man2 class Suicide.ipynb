```python
explode=[0,0.1]
plt.figure(figsize=(8,10))
master['sex'].value_counts().plot(kind='pie', autopct='%.2f%%', explode=explode, colors=['red', 'yellow'])
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.catplot(x='generation', hue='sex', kind='count', data=p , height=10, palette='muted')
plt.show()
```

Since Males and Females are in same percentage, looks like suicides are majorly done by "Generation X" and "Silent" and interesting part is Millenials(which is present Generation) is in 3rd Position
